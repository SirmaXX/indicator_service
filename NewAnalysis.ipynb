```python
import pandas as pd
from pandas_profiling import profile_report

from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from mlxtend.feature_selection import SequentialFeatureSelector

data = pd.read_csv("dataset/diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
data.head()
```

```python
X_train, X_val, y_train, y_val = train_test_split(data, data['Diabetes_binary'], test_size=0.20, random_state=123)
y_train = y_train.values.ravel()
y_val = y_val.values.ravel()
```

```python
sfs = SequentialFeatureSelector(linear_model.LogisticRegression(),
                                k_features=3,
                                forward=True,
                                scoring='accuracy',
                                cv=None)
selected_features = sfs.fit(X_train, y_train)
```

```python
logreg = linear_model.LogisticRegression()
logreg.fit(X_train, y_train)
 
# Make predictions using the test set
y_pred = logreg.predict(X_val)
 
# Evaluate the model performance
print(y_pred)
```
